```python
import pandas as pd
import matplotlib.pyplot as plt
import pickle
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv("breast_cancer.csv")
df.head(20)
```

```python
df.Class.unique()
```

```python
x_attributes = df.drop(['Class'] ,axis = 'columns')
x_attributes
```

```python
y_attribute = df.Class
y_attribute
```

```python
X_train, X_test, y_train, y_test = train_test_split(x_attributes, y_attribute, test_size=0.1, random_state=42)
```

```python
X_train
```

```python
X_test
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
```

```python
model.fit(X_train,y_train)
```

```python
model.predict(X_test)
```

```python
model.score(X_test,y_test)
```

```python
model.predict([[5,1,1,1,2,1,3,1,1]])
```

```python
with open("Saved_model" , "wb" ) as file:
    pickle.dump(model,file)
```
